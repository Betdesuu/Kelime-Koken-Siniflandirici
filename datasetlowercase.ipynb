```python
import pandas as pd
```

```python
df = pd.read_csv("final_dataset_no_conflicts.csv")
```

```python
def turkish_lower(text):
    text = str(text)
    text = text.replace("I", "ı").replace("İ", "i")
    return text.lower().strip()
```

```python
df["kelime"] = df["kelime"].apply(turkish_lower)
```

```python
df.to_csv("final_dataset_lowercase.csv", index=False, encoding="utf-8")
```
